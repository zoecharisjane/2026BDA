## numpy.hstack()

```python
import numpy as np

a = np.arange(6).reshape(3, 2)
a
```

```python
b = np.arange(10, 19).reshape(3, 3)
b
```

```python
c = np.array([[10, 20], [30, 40]])
c
```

```python
# np.hstack((a, c)) # Original code caused ValueError due to mismatch in number of rows

# To fix, ensure arrays have the same number of rows.
# Here, we slice 'a' to match the number of rows in 'c'.
np.hstack((a[:2], c))
```

```python
np.hstack((np.arange(3), np.arange(10, 15)))
```
