## 리스트 컴프리헨션

```
my_list = [x for x in range(10)]
my_list
```

```
[x for x in range(10) if x % 2 == 1]
```

```
['even' if i % 2 == 0 else 'odd' for i in range(1, 11)]
```

```
[[i for j in range(5)] for i in range(3)]
```

```
[[k for k in range(i, i+6)] for i in range(3)]
```
